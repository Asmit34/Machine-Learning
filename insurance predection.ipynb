```python
auto=df[['age','sex','bmi','children','smoker','region','expenses']]
auto

```

```python
figsize=(13,13)
sns.pairplot(auto)
plt.show()
```

```python
gender=pd.get_dummies(auto["sex"],drop_first=True)
auto=pd.concat([auto,gender],axis=1)
auto

```

```python
auto.head()
```

```python
smoker=pd.get_dummies(auto["smoker"],drop_first=True)
auto=pd.concat([auto,smoker],axis=1)
auto
```

```python
region_typ=pd.get_dummies(auto["region"],drop_first=True)
auto=pd.concat([auto,region_typ],axis=1)
auto
```

```python
auto.drop(["sex","region","smoker"],axis=1,inplace=True)
auto
```

```python
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
df_train,df_test=train_test_split(auto,train_size=0.7,test_size=0.3,random_state=100)
```

```python
df_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
#applying scaler() to all the column except dummy variable
num_vars=['age','bmi','children','expenses']
df_train[num_vars]=scaler.fit_transform(df_train[num_vars])
```

```python
df_train.describe()
```

```python
df_train.head()     #dimension reduction
```

```python
#let's check correlection coeff. to see which variable is highly correlated
plt.figure(figsize=(15,9))
sns.heatmap(df_train.corr(),annot=True)
plt.show()
```

```python
y_train = df_train.pop('expenses')
X_train = df_train
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

#rfe = RFE(lm, 10)
rfe = RFE(lm, n_features_to_select=10)
#running RFE
rfe = rfe.fit(X_train, y_train)
rfe
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col=X_train.columns[rfe.support_]
col
```

```python
X_train_rfe=X_train[col]
X_train_rfe
```

```python
import statsmodels.api as sm
X_train_rfe=sm.add_constant(X_train_rfe)
```

```python
X_train_rfe
```

```python
lm=sm.OLS(y_train,X_train_rfe).fit()  #running the linear model
print(lm.summary())
```

```python
#dropping male as p value is high
X_train_new=X_train_rfe.drop(["male"],axis=1)
X_train_new
```

```python
X_train_lm=sm.add_constant(X_train_new)
X_train_lm
```

```python
lm=sm.OLS(y_train,X_train_lm).fit()
lm.summary()
```

```python
#Dropping the northeast as p value is high
X_train_new1=X_train_new.drop(["northwest"],axis=1)
X_train_new1
```

```python
import statsmodels.api as sm
X_train_lm=sm.add_constant(X_train_new1)
lm=sm.OLS(y_train,X_train_lm).fit()
print(lm.summary())
```

```python
X_train_new2=X_train_new1.drop(["southeast"],axis=1)
X_train_new2
```

```python
X_train_lm=sm.add_constant(X_train_new2)
lm=sm.OLS(y_train,X_train_lm).fit()
lm.summary()
```

```python
X_train_new3=X_train_new2.drop(["southwest"],axis=1)
X_train_new3
```

```python
X_train_lm=sm.add_constant(X_train_new3)
lm=sm.OLS(y_train,X_train_lm).fit()
lm.summary()
```

```python
X_train_new4=X_train_new3.drop(["children"],axis=1)
X_train_new4
```

```python
X_train_lm=sm.add_constant(X_train_new4)
lm=sm.OLS(y_train,X_train_lm).fit()
lm.summary()
```
