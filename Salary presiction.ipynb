```python
auto=df[['Age','Gender','Education Level','Years of Experience','Salary']]
auto
```

```python
sns.pairplot(auto)
plt.show()
```

```python
df.head()
```

```python
Sex=pd.get_dummies(auto['Gender'],drop_first=True)
auto=pd.concat([auto,Sex],axis=1)
auto
```

```python
Education_level=pd.get_dummies(auto['Education Level'])
auto=pd.concat([auto,Education_level],axis=1)
auto
```

```python
auto.drop(['Gender','Education Level'],axis=1,inplace=True)
```

```python
auto
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test=train_test_split(auto,train_size=0.7,test_size=0.3,random_state=100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
num_vars=["Age",'Years of Experience','Salary']
df_train[num_vars]=scaler.fit_transform(df_train[num_vars])
```

```python
df_train
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(df_train.corr(),annot=True)
plt.show()
```

```python
y_train=df_train.pop("Salary")
x_train=df_train
```

```python
y_train
```

```python
x_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
rfe=RFE(lm,n_features_to_select=10 )
rfe =rfe.fit(x_train,y_train)
rfe
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
selected_column=x_train.columns[rfe.support_]
selected_column
```

```python
selected_column_rfe=x_train[selected_column]
selected_column_rfe
```

```python
import statsmodels.api as sm
selected_column_rfe=sm.add_constant(selected_column_rfe)
selected_column_rfe
```

```python
lm=sm.OLS(y_train,selected_column_rfe).fit()
print(lm.summary())
```

```python
x_train_new=selected_column_rfe.drop(["Master's"],axis=1)
x_train_new
```

```python
x_train_lm=sm.add_constant(x_train_new)
x_train_lm
```

```python
lm=sm.OLS(y_train,x_train_lm).fit()
print(lm.summary())
```

```python
x_train_new1=x_train_lm.drop(["PhD"],axis = 1)
x_train_new1
```

```python
x_train_lm=sm.add_constant(x_train_new1)
x_train_lm
```

```python
lm=sm.OLS(y_train,x_train_lm).fit()
print(lm.summary())
```

```python
#Salary  = Age*0.4112 + Years of Experience + Male*0.0460 + Bachelor's*(-0.0918) - 0.0237
```

```python
y_train_price = lm.predict(x_train_lm)
y_train_price
```

```python
num_vars = ['Age','Years of Experience','Salary']
df_test[num_vars] = scaler.transform(df_test[num_vars])
print(df_test[num_vars])
```

```python
y_test = df_test.pop('Salary')
X_test = df_test
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[['Age','Years of Experience','Male',"Bachelor's"]]

# Adding a constant variable
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_pred = lm.predict(X_test_new)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)                          # Y-label
```
