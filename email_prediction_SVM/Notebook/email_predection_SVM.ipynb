```python
import string
from nltk.corpus import stopwords

def text_process(mess):
    # Remove punctuation
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)

    # Remove stopwords
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
data["v2"].apply(text_process)
```

```python
from sklearn.model_selection import train_test_split
email_train,email_test,label_train,label_test=train_test_split(data['v2'],data['v1'],test_size=0.2,shuffle=True)
print(len(email_train),len(email_test),len(email_train)+len(email_test))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
pipeline = Pipeline([
    ('bow', CountVectorizer(analyzer=text_process)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', SVC(probability=True)),  # train on TF-IDF vectors w/ SVM
])
```

```python
pipeline.fit(email_train,label_train)
```

```python
# Save the pipeline to a file
joblib.dump(pipeline, 'email_prediction_model.pkl')
```

```python
predictions = pipeline.predict(email_test)
predictions
```

```python
prediction = pipeline.predict(email_train)
training_acc = accuracy_score(prediction, label_train)
print("Accuracy on training data:", training_acc)
```

```python
# accuracy on test data
preyd = pipeline.predict(email_test)
testing_acc = accuracy_score(preyd, label_test)
print("Accuracy on testing data:",testing_acc)
```

```python
#Classification report
len(pipeline.predict(email_train))
print(classification_report(label_test, pipeline.predict(email_test)))
# print(confusion_matrix(label_test,model.predict(email_test)))
```

```python
cm = confusion_matrix(label_test,pipeline.predict(email_test))
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True, fmt='.2f', linewidths=.7, square=True,cmap='Blues_r')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(accuracy_score(label_test,pipeline.predict(email_test)))
plt.title(all_sample_title,size=15)
```
